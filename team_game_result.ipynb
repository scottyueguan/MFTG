```
from __future__ import print_function




import numpy as np
from scipy.stats import rv_discrete
from scipy.stats import entropy
from collections import deque
import math
import random
from copy import deepcopy
import matplotlib.pyplot as plt
import pickle
from scipy.special import softmax





import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.autograd import Variable




class mamdp_model:

    def __init__(self,team1_size,team2_size,state_space,action_space,transition_probability,observation_space,observation_probability,cost,discount_factor):

        """
        Each team can have multiple players. The id of each player is stored as a list. Players' ids in each team must be integers starting from 0 to size of the team.
        """

        self.t1_size = team1_size
        self.t2_size = team2_size


        """
        The state space is a list. The order of states must be respected because it is used in the transition probability matrix.
        """
        self.state_space = state_space



        """
        Action spaces: each player in each team has an action space.
        """


        self.action_space = action_space



        """
        The transition probabilities are representedas a dictionary with action as the key and the associated transition as the value.
        """


        self.transition_probability = transition_probability


        """
        Observation space, size and probabilities: observation space is a list. Observation probability is a dictionary with state as the key and observation distribution as the value.
        """
        self.observation_space = observation_space

        self.observation_probability = observation_probability

        """
        Instantaneous costs are represented as dictionary. The key is of the form state,action and its value is the cost.
        """
        self.cost = cost


        "Discount factor is in [0,1]"
        self.discount_factor = discount_factor




states = ['0','1','2','3']
statesize = len(states)
actions = [[['a','b'],['a','b']],[['a','b','c']]]



pshift = 0.00
pvan_al = 0.3
pvan_non = 0.3
pattack = 0.3
pdisrupt = 0.5
# trans = {'aaa':np.array([[1-pshift,pshift,0,0],[0,0,0,1],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'aab':np.array([[0,0,1,0],[0,1,0,0],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'aac':np.array([[(1-pshift)*(1-pvan_al),pshift*(1-pvan_al),pvan_al,0],[0,1-pvan_non,0,pvan_non],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'aba':np.array([[1,0,0,0],[0,0,0,1],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'abb':np.array([[0,0,1,0],[0, 1,0,0],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'abc':np.array([[1-pvan_non,0,pvan_non,0],[pshift*(1-pvan_al),(1-pshift)*(1-pvan_al),0,pvan_al],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'baa':np.array([[1-pshift,pshift,0,0],[0,0,0,1],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'bab':np.array([[0,0,1,0],[0,1,0,0],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'bac':np.array([[(1-pshift)*(1-pvan_al),pshift*(1-pvan_al),pvan_al,0],[0,1-pvan_non,0,pvan_non],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'bba':np.array([[1,0,0,0],[0,0,0,1],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'bbb':np.array([[0,0,1,0],[0,1,0,0],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'bbc':np.array([[1-pvan_non,0,pvan_non,0],[pshift*(1-pvan_al),(1-pshift)*(1-pvan_al),0,pvan_al],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]])}
trans = {'aaa':np.array([[1-pdisrupt,pdisrupt,0,0],[0,0,0,1],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'aab':np.array([[0,0,1,0],[pdisrupt,1-pdisrupt,0,0],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'aac':np.array([[(1-pshift)*(1-pvan_al),pshift*(1-pvan_al),pvan_al,0],[0,1-pvan_non,0,pvan_non],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'aba':np.array([[1-pdisrupt,pdisrupt,0,0],[0,0,0,1],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'abb':np.array([[0,0,1,0],[pdisrupt,1-pdisrupt,0,0],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'abc':np.array([[1-pvan_non,0,pvan_non,0],[0,(1-pvan_al),0,pvan_al],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'baa':np.array([[1-pdisrupt,pdisrupt,0,0],[0,0,0,1],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'bab':np.array([[0,0,1,0],[pdisrupt,1-pdisrupt,0,0],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'bac':np.array([[(1-pvan_al),0,pvan_al,0],[0,1-pvan_non,0,pvan_non],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'bba':np.array([[1-pdisrupt,pdisrupt,0,0],[0,0,0,1],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'bbb':np.array([[0,0,1,0],[pdisrupt,1-pdisrupt,0,0],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]]),'bbc':np.array([[1-pvan_non,0,pvan_non,0],[pshift*(1-pvan_al),(1-pshift)*(1-pvan_al),0,pvan_al],[pattack/2,pattack/2,1-pattack,0],[pattack/2,pattack/2,0,1-pattack]])}
obspace = [[['0','1'],['0','1']],['0','1']]
# obsize = len(obspace)
# q0 = 0.1
# q1 = 0.5
obsprob = [[{'0': np.array([1,0]), '1': np.array([1,0]), '2': np.array([0,1]), '3': np.array([0,1])},{'0': np.array([1,0]), '1': np.array([1,0]), '2': np.array([0,1]), '3': np.array([0,1])}],[{'0': np.array([1,0]), '1': np.array([1,0]), '2': np.array([0,1]), '3': np.array([0,1])}]]
#
# ytransform = {}
# count = 0
# for y in obspace:
#     tempvec = []
#     for x in states:
#         tempvec.append(obsprob[x][count])
#     ytransform[y] = tempvec
#     count = count + 1

#ytransform = {'0': [1.0,0.0], '1': [0.0,1.0]}

# reward = {'0,a': 20.0, '0,b': 0.0, '1,a': 5.0, '1,b':15.0}
# rewardvec = {'a': np.array([20.0,0.0]).reshape(1,2), 'b': np.array([5.0,15.0]).reshape(1,2)}


acost = 15  # Cost inflicted by a successful targeted attack
bcost = 20  # Cost inflicted by an undefended blanket attack
bcost_less = 10 # Cost inflicted by a blanket attack with defense
costvec = {'aaa': np.array([acost,0,0,0]), 'aab': np.array([0,acost,0,0]), 'aac': np.array([bcost_less,bcost,0,0]),'aba': np.array([acost,0,0,0]), 'abb': np.array([0,acost,0,0]), 'abc': np.array([bcost,bcost_less,0,0]),'baa': np.array([acost,0,0,0]), 'bab': np.array([0,acost,0,0]), 'bac': np.array([bcost_less,bcost,0,0]),'bba': np.array([acost,0,0,0]), 'bbb': np.array([0,acost,0,0]), 'bbc': np.array([bcost,bcost_less,0,0])}

disc = 0.9


model = mamdp_model(2,1,states,actions,trans,obspace,obsprob,costvec,disc)

"""
Neural network that represents the value function
"""


class value_net(nn.Module):

    def __init__(self,input_size,zero_init):
        super().__init__()

        h1_in = 10
        h1_out = 10
        self.layer1 = nn.Linear(input_size,h1_in, bias=False)
        self.layer2 = nn.Linear(h1_in,h1_out,bias=False)
        self.layer3 = nn.Linear(h1_out,1,bias=False)
        if zero_init:
            torch.nn.init.zeros_(self.layer1.weight)
            torch.nn.init.zeros_(self.layer2.weight)
            torch.nn.init.zeros_(self.layer3.weight)

    def forward(self,bel):
        bel_int = F.leaky_relu(self.layer1(bel))
        bel_int = F.leaky_relu(self.layer2(bel_int))
        return self.layer3(bel_int)


class dist_net(nn.Module):

    def __init__(self,dist_size):
        super().__init__()
        self.layer1 = nn.Linear(1,dist_size, bias=False)

    def forward(self,x):
        return F.softmax(self.layer1(x))


class maxcost_net(nn.Module):

    def __init__(self,valnet):
        super().__init__()

        self.layers = nn.ModuleDict({})

        self.layers['pres000'] = dist_net(2)
        self.layers['pres001'] = dist_net(2)
        self.layers['pres002'] = dist_net(2)
        self.layers['pres003'] = dist_net(2)
        self.layers['pres01'] = dist_net(2)

        self.layers['val_next'] = value_net(4,False)
        params1 = self.layers['val_next'].named_parameters()
        params2 = valnet.named_parameters()


        dict_params1 = dict(params1)

        # print(dict_params1)
        for name1, param1 in params2:
            if name1 in dict_params1:
                dict_params1[name1].data.copy_(param1.data)

        # print('Params')
        # params3 = self.layers['val_next'].named_parameters()
        # print(dict(params3))


    def forward(self,bel,mamdp_model):
        advact = mamdp_model.action_space[1][0]
        defact = mamdp_model.action_space[0][0]
        states = mamdp_model.state_space
        obspace = mamdp_model.observation_space[0][0]
        obsprob = mamdp_model.observation_probability[0][0]
        statesize = len(states)
        delta = mamdp_model.discount_factor
        dummytensor = torch.tensor([1],dtype=torch.float)

        re_bel = torch.reshape(bel,(-1,statesize,1))
        preslist = []
        for s in states:
            preslist.append(self.layers['pres00'+s].forward(dummytensor))
        pres00 = torch.stack(preslist)

        multemp = torch.mul(re_bel,pres00)
        multemp = torch.transpose(multemp,-1,-2)

        pres01 = self.layers['pres01'].forward(dummytensor)
        pres01 = torch.reshape(pres01,(-1,1,1))

        multemp = torch.mul(multemp,pres01)




        advactiter = iter(advact)
        a2 = advact[0]
        costlist = []
        for a1 in defact:
            templist = []
            for a0 in defact:
                templist.append(mamdp_model.cost[a0+a1+a2])
            costlist.append(deepcopy(templist))
        costensor = torch.tensor(costlist)
        maxtemp = torch.sum(torch.mul(multemp,costensor))

        county = 0
        for y in obspace:
            obsvec = np.zeros(len(states))
            countx = 0
            for x in states:
                obsvec[countx] = obsprob[x][county]
                countx += 1
            county += 1
            count1 = 0
            for a1 in defact:
                count2 = 0
                for a0 in defact:
                    count3 = 0
                    temptensor = torch.zeros(len(states))
                    for s in states:
                        temptensor += multemp[count1,count2,count3]*torch.FloatTensor(np.multiply(mamdp_model.transition_probability[a0+a1+a2][count3],obsvec))
                        count3 += 1
                    maxtemp += delta*torch.sum(self.layers['val_next'](temptensor))
                    count2 += 1
                count1 += 1
        # print('addTemp')
        # print(maxtemp)



        next(advactiter)
        for a2 in advactiter:
            costlist = []
            for a1 in defact:
                templist = []
                for a0 in defact:
                    templist.append(mamdp_model.cost[a0+a1+a2])
                costlist.append(deepcopy(templist))
            costensor = torch.tensor(costlist)
            addtemp = torch.sum(torch.mul(multemp,costensor))




            county = 0
            for y in obspace:
                obsvec = np.zeros(len(states))
                countx = 0
                for x in states:
                    obsvec[countx] = obsprob[x][county]
                    countx += 1
                county += 1
                count1 = 0
                for a1 in defact:
                    count2 = 0
                    for a0 in defact:
                        count3 = 0
                        temptensor = torch.zeros(len(states))
                        for s in states:
                            temptensor += multemp[count1,count2,count3]*torch.FloatTensor(np.multiply(mamdp_model.transition_probability[a0+a1+a2][count3],obsvec))
                            count3 += 1
                        addtemp += delta*torch.sum(self.layers['val_next'](temptensor))
                        count2 += 1
                    count1 += 1
                # print('addTemp')
                # print(addtemp)

            maxtemp = torch.max(maxtemp,addtemp)

        # print('maxTemp')
        # print(maxtemp)
        return maxtemp



def minmax(bel,mcnet,model):
    states = model.state_space

    val = float('inf')
    presat0 = 0
    presat1 = 0
    presat2 = 0
    presat3 = 0
    presat4 = 0



    n_epochs = 4000
    params = []
    for s in states:
        params += list(mcnet.layers['pres00'+s].parameters())
    params += list(mcnet.layers['pres01'].parameters())


    for attempt in range(20):
        for param in params:
            torch.nn.init.normal_(param)

        minmax_optimizer = optim.Adam(params,lr = 0.1)

        prevout = 1000

        for ep in range(n_epochs):

            output = mcnet.forward(bel,model)
            # print(ep)
            # print('minmax_out')
            # # print(bel)
            # # print(output)
            # print(params)
            if abs(output.item()-prevout) < 0.001:
                break
            prevout = output.item()

            output.backward()
            minmax_optimizer.step()
            minmax_optimizer.zero_grad()

        if output.item() < val:
            presat0 = softmax(params[0].data.numpy())[0][0]
            presat1 = softmax(params[1].data.numpy())[0][0]
            presat2 = softmax(params[2].data.numpy())[0][0]
            presat3 = softmax(params[3].data.numpy())[0][0]
            presat4 = softmax(params[4].data.numpy())[0][0]
        val = min(val,output.item())

    print("Prescriptions")
    print(presat0)
    print(presat1)
    print(presat2)
    print(presat3)
    print(presat4)


    return val


def valnet_fit(mcnet,model,tol):
    belresol = 0.01
    belrange = np.arange(0,1+belresol,belresol)
    nobelres = len(belrange)
    inplist = []
    newvalnet = value_net(4,False)
    fit_params = newvalnet.parameters()
    for p in belrange:
        inplist.append([1-p,p,0,0])
    for p in belrange:
        inplist.append([0,0,1-p,p])

    # count = 0
    # value = np.zeros(nobelres)

    target = []
    for inp in inplist:
        print("Beliefs")
        print(inp)
        temptensor = torch.tensor(inp)
        target.append(minmax(temptensor,mcnet,model))


    inputs = torch.FloatTensor(inplist)
    # print(inputs)
    targets = torch.tensor(target)
    targets = targets.view(len(inplist), -1)
    # print(targets)

    optimizer = optim.Adam(fit_params,lr=0.1)
    criterion = nn.MSELoss()

    n_epochs = 50000
    for ep in range(n_epochs):
        output = newvalnet(inputs)

        loss = criterion(output,targets)
        
        if (loss.item() < tol):
            break
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
    print('Loss:', loss.item())
    return [newvalnet,target]




horizon = 15

valnetlist = []
costnetlist = []
valnetlist.append([value_net(4,True),0])

for t in range(horizon):
    params = list(valnetlist[0][0].parameters())
    costnetlist.append(maxcost_net(valnetlist[t][0]))
    valnetlist.append(valnet_fit(costnetlist[t],model,0.0005*(1)))
    plotresolution = 0.01
    prange = np.arange(0,1+plotresolution,plotresolution)
    noplotres = len(prange)
    valuefunc = np.zeros(noplotres)


    valuefunc_est = np.zeros(noplotres)
    count = 0

    for p in prange:
        print(p)
        input = torch.FloatTensor([1-p,p,0,0])
        valuefunc[count] = valnetlist[t+1][1][count]
        valuefunc_est[count] = valnetlist[t+1][0](input).item()
        count += 1



    fig1 = plt.figure()
    ax = plt.subplot(111)

    ax.plot(prange, valuefunc, label = 'Value function',linewidth = 2,linestyle='-', marker="x", markersize = 1, markevery = 2)
    ax.plot(prange, valuefunc_est, label = 'Estimated Value function',linewidth = 2,linestyle='-', marker="o", markersize = 1, markevery = 2)

    ax.legend(loc='best')
    # plt.show()
    fig1.savefig('fig_'+str(t+1)+'.eps')




# valnet = value_net(4)
#
# mcnet = maxcost_net(valnet)
#
#
# res = 0.01
# prange = np.arange(0,1+res,res)
# nobel = len(prange)
#
# bel_state = torch.tensor([0.5,0.5,0,0])
# print(valnet(bel_state))
#
# out = mcnet.forward(bel_state,model)
#
# print(minmax(bel_state,mcnet,model))
# newvalnet = valnet_fit(mcnet,model)




"""
Plot value functions
"""


# plotresolution = 0.01
# prange = np.arange(0,1+plotresolution,plotresolution)
# noplotres = len(prange)
# valuefunc = np.zeros(noplotres)


# valuefunc_est = np.zeros(noplotres)
# count = 0

# for p in prange:
#     print(p)
#     input = torch.FloatTensor([1-p,p,0,0])
#     valuefunc[count] = valnetlist[horizon][1][count]
#     valuefunc_est[count] = valnetlist[horizon][0](input).item()
#     count += 1

# print(valuefunc)



# fig1 = plt.figure()
# ax = plt.subplot(111)

# ax.plot(prange, valuefunc, label = 'Value function',linewidth = 2,linestyle='-', marker="x", markersize = 1, markevery = 2)
# ax.plot(prange, valuefunc_est, label = 'Estimated Value function',linewidth = 2,linestyle='-', marker="o", markersize = 1, markevery = 2)

# ax.legend(loc='best')
# plt.show()
# fig1.savefig('fig_2.eps')
```
